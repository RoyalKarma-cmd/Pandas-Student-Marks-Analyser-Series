```python
import pandas as pd
import numpy as np
```

```python
print("=" * 50)
print("STUDENT MARKS ANALYSIS")
print("=" * 50)

marks = pd.Series([95,88,68,65,35,60,75])

print('\nOriginal Marks:-')
print("-" * 30)
print(marks)

print('\nStatistical Report')
print("-" * 30)
print("Average Marks:", marks.mean())
print("Highest Marks:", marks.max())
print("Lowest Marks:", marks.min())
print("Median marks", marks.median())

print('\nPerformance Report')
print("-" * 30)

topper_Students = marks[marks > 80]

print("\nDistinction Students:-")
print(topper_Students)

average_Students = marks[(marks > 50) & (marks < 80)]

print("\nAverage Students:-")
print(average_Students)

passed_students = marks[marks >= 45]

print("\nPassed Students")
print(passed_students)

failed_students = marks[marks < 45]

print("\nFailed Students")
print(failed_students)

print('\nPass/Fail Report')
print("-" * 30)
result = np.where(
    marks >= 45,
    "Pass",
    "Fail"
)
print(result)

print('\nGrade Report:-')
print("-" * 30)

for student in marks:

    if student >= 90:
        grade = "A+"
    
    elif student >= 80:
        grade = "A"
    
    elif student >= 70:
        grade = "B"

    elif student >= 60:
        grade = "C"
    
    elif student >= 50:
        grade = "D"

    else:
        grade = "F"

    print(f"Marks: {student}  --> Grade: {grade}")

print("\nSummary")
print("-" * 30)

print("Total Students:-", marks.size)
print("Passed Students:-", passed_students.size)
print("Failed Students:-", failed_students.size)
print("Top performers:-", topper_Students.size)
print("Average Performer:-", average_Students.size)


```
